```python
"""
Coordenadas dos Bairros de Araraquara/SP - VERS√ÉO ATUALIZADA
Coordenadas corrigidas e complementadas
Data: Janeiro 2026
"""

# Coordenadas aproximadas do centro de cada bairro
# Araraquara est√° aproximadamente em: -21.7844, -48.1756

BAIRROS_ARARAQUARA = {
    # ============================================================
    # ZONA CENTRAL
    # ============================================================
    'Centro': (-21.7844, -48.1756),
    'Centro Hist√≥rico': (-21.7844, -48.1756),
    
    # ============================================================
    # ZONA NORTE (Vila Xavier e adjac√™ncias)
    # ============================================================
    'Vila Xavier': (-21.7650, -48.1650),
    'Vila Xavier (Vila Xavier)': (-21.7650, -48.1650),
    
    'Jardim Am√©rica': (-21.7750, -48.1700),
    'Jardim Am√©rica (Vila Xavier)': (-21.7750, -48.1700),
    
    'Jardim Paulista': (-21.7650, -48.1650),
    'Jardim Paulistano (Vila Xavier)': (-21.7750, -48.1600),
    
    'Jardim Pinheiros': (-21.7697, -48.1364),
    'Jardim Pinheiros (Vila Xavier)': (-21.7697, -48.1364),
    'Parque Pinheiros': (-21.7600, -48.1720),
    
    'Jardim Brasil': (-21.7844, -48.1540),
    'Jardim Brasil (Vila Xavier)': (-21.7844, -48.1540),
    
    'Jardim Bras√≠lia (Vila Xavier)': (-21.7671, -48.1565),
    
    'Vila Biagioni (Vila Xavier)': (-21.7714, -48.1411),
    
    'Jardim Morada do Sol (Vila Xavier)': (-21.7750, -48.1599),
    
    'Jardim Europa (Vila Xavier)': (-21.7700, -48.1550),
    
    'Jardim Viaduto (Vila Xavier)': (-21.7680, -48.1580),
    
    'Vila Esperan√ßa (Vila Xavier)': (-21.7720, -48.1620),
    
    'Vila Renata (Vila Xavier)': (-21.7690, -48.1590),
    
    'Jardim Santa Clara (Vila Xavier)': (-21.7730, -48.1610),
    
    'Vila Tito de Carvalho (Vila Xavier)': (-21.7700, -48.1570),
    
    'Jardim Tabapu√£ (Vila Xavier)': (-21.7660, -48.1640),
    
    'Parque Gramado (Vila Xavier)': (-21.7710, -48.1630),
    
    'Jardim Floridiana (Vila Xavier)': (-21.7740, -48.1650),
    
    'Vila Vieira (Vila Xavier)': (-21.7680, -48.1600),
    
    'Jardim Silv√¢nia (Vila Xavier)': (-21.7720, -48.1660),
    
    'Vila Santa Maria (Vila Xavier)': (-21.7760, -48.1620),
    
    # ============================================================
    # ZONA SUL
    # ============================================================
    'Vila Harmonia': (-21.7668, -48.1736),
    'Vila Harmonia I': (-21.7950, -48.1800),
    'Vila Harmonia II': (-21.7980, -48.1820),
    
    'Jardim Higien√≥polis': (-21.8000, -48.1750),
    
    'Jardim Bandeirantes': (-21.7805, -48.2243),
    
    'Jardim das Flores': (-21.7885, -48.2061),
    
    'Jardim dos Manac√°s': (-21.7856, -48.1990),
    
    'Jardim Tropical': (-21.8050, -48.1650),
    
    'Jardim Nova Araraquara': (-21.7750, -48.2051),
    
    # ============================================================
    # ZONA LESTE
    # ============================================================
    'Parque Laranjeiras': (-21.7893, -48.1999),
    
    'Jardim Imperador': (-21.7546, -48.1595),
    'Jardim Imperador II': (-21.7546, -48.1595),
    
    'Jardim S√£o Francisco': (-21.7461, -48.1569),
    
    'Parque Residencial Iguatemi': (-21.7880, -48.1620),
    'Jardim Iguatemi': (-21.7900, -48.1620),
    
    'Jardim Residencial Para√≠so': (-21.7900, -48.1650),
    
    'Jardim S√£o Rafael I': (-21.7457, -48.1397),
    'Jardim S√£o Rafael II': (-21.7430, -48.1426),
    
    'Jardim Indai√°': (-21.7465, -48.1433),
    
    'Jardim Aclima√ß√£o': (-21.7587, -48.1558),
    
    'Jardim Adalgisa': (-21.7591, -48.1618),
    
    'Jardim Bounganville': (-21.7544, -48.1654),
    
    'Jardim Margarida': (-21.7564, -48.1651),
    
    'Jardim S√£o Jorge': (-21.7740, -48.1611),
    
    # ============================================================
    # ZONA OESTE
    # ============================================================
    'Cidade Jardim': (-21.7472, -48.1878),
    
    'Vila Ferrovi√°ria': (-21.7767, -48.1715),
    
    'Vila Suconasa': (-21.7780, -48.1950),
    
    'Campus ville': (-21.7750, -48.1880),
    
    'Parque S√£o Paulo': (-21.7900, -48.1900),
    'Parque S√£o Paulo (Vila Xavier)': (-21.7813, -48.1304),
    
    'Selmi Dei': (-21.7950, -48.1950),
    'Jardim Roberto Selmi Dei': (-21.7413, -48.1522),
    
    'Jardim Dona Zulmira': (-21.7850, -48.1950),
    
    'Vila Velosa': (-21.7677, -48.1675),
    
    # ============================================================
    # ZONA INDUSTRIAL
    # ============================================================
    'Vila Cidade Industrial': (-21.7700, -48.1550),
    'Vila Cidade Industrial (Vila Xavier)': (-21.7700, -48.1550),
    
    'Distrito Industrial': (-21.7600, -48.1500),
    'Parque Industrial': (-21.7650, -48.1500),
    
    'Jardim dos Industri√°rios': (-21.8242, -48.1439),
    
    # ============================================================
    # OUTROS BAIRROS (ORDEM ALFAB√âTICA)
    # ============================================================
    'Altos do Jaragu√°': (-21.7795, -48.1985),
    
    'Ch√°cara Flora Araraquara': (-21.7266, -48.1821),
    
    'Doutor Tancredo de Almeida Neves': (-21.7820, -48.1680),
    
    'Estrela e Kanashiro': (-21.7880, -48.1920),
    
    'Jardim Adalberto Frederico de Oliveira Roxo I': (-21.7850, -48.1450),
    
    'Jardim √Åguas do Paiol': (-21.7620, -48.2071),
    
    'Jardim Almeida': (-21.7750, -48.1580),
    
    'Jardim Altos do Cecap I e II': (-21.7920, -48.1480),
    
    'Jardim Ana Adelaide': (-21.7770, -48.1690),
    
    'Jardim Arang√°': (-21.7950, -48.1850),
    
    'Jardim Araraquara': (-21.7750, -48.2051),
    
    'Jardim √Årtico': (-21.7720, -48.1620),
    
    'Jardim Boa Vista': (-21.7880, -48.1780),
    
    'Jardim Bot√¢nico': (-21.7573, -48.1817),
    
    'Jardim Calif√≥rnia': (-21.8007, -48.1943),
    
    'Jardim Cruzeiro do Sul I e II': (-21.8100, -48.1650),
    
    'Jardim das Esta√ß√µes (Vila Xavier)': (-21.7727, -48.1496),
    
    'Jardim das Roseiras': (-21.7850, -48.2000),
    
    'Jardim do Bosque': (-21.7880, -48.2050),
    
    'Jardim do Carmo': (-21.7920, -48.1880),
    
    'Jardim Domingos S√°vio': (-21.8050, -48.1520),
    
    'Jardim dos Ip√™s': (-21.7950, -48.2100),
    
    'Jardim dos Oitis': (-21.7890, -48.2020),
    
    'Jardim Eliana': (-21.7800, -48.1650),
    
    'Jardim Esplanada': (-21.8178, -48.1453),
    
    'Jardim Gard√™nias': (-21.7920, -48.1920),
    
    'Jardim Imperial': (-21.8183, -48.1432),
    
    'Jardim Magn√≥lias': (-21.7870, -48.2010),
    
    'Jardim Mangiacapra': (-21.7950, -48.1950),
    
    'Jardim Maria Luiza': (-21.7256, -48.0977),
    'Jardim Maria Luiza V': (-21.7421, -48.1976),
    
    'Jardim Marialice': (-21.7780, -48.1720),
    
    'Jardim Marivan': (-21.7860, -48.1940),
    
    'Jardim Morumbi': (-21.7680, -48.1580),
    
    'Jardim Nova Am√©rica': (-21.7820, -48.2080),
    
    'Jardim Nova √âpoca': (-21.8120, -48.1580),
    
    'Jardim Padre Anchieta': (-21.7900, -48.1750),
    
    'Jardim Panorama': (-21.7950, -48.2000),
    
    'Jardim Primavera': (-21.7850, -48.2100),
    
    'Jardim Primor': (-21.7880, -48.1960),
    
    'Jardim Quitandinha': (-21.8050, -48.1600),
    'Jardim Quitandinha II': (-21.8070, -48.1620),
    
    'Jardim Residencial Elvio Lupo': (-21.8100, -48.1700),
    
    'Jardim Residencial Iedda': (-21.8158, -48.1395),
    
    'Jardim Residencial It√°lia': (-21.7950, -48.2050),
    
    'Jardim Residencial Silvestre': (-21.8257, -48.1413),
    
    'Jardim Santa Angelina': (-21.7980, -48.2030),
    
    'Jardim Santa J√∫lia (Vila Xavier)': (-21.7927, -48.1501),
    'Jardim Santa J√∫lia III': (-21.7940, -48.1520),
    
    'Jardim Santa L√∫cia': (-21.6881, -48.1319),
    
    'Jardim Santa Marta': (-21.8191, -48.1490),
    
    'Jardim Santa Rita de C√°ssia': (-21.8020, -48.1680),
    
    'Jardim S√£o Gabriel': (-21.7917, -48.1978),
    
    'Jardim Tamoio': (-21.8137, -48.1793),
    
    'Jardim Tinen': (-21.7850, -48.1550),
    
    'Jardim Uirapuru': (-21.8000, -48.2100),
    
    'Jardim Universal': (-21.7934, -48.2065),
    
    'Jardim Vale das Rosas': (-21.7990, -48.2020),
    
    'Jardim Vit√≥ria': (-21.7920, -48.1980),
    
    'Parque Alvorada': (-21.7850, -48.2150),
    
    'Parque Atlanta': (-21.7792, -48.2096),
    
    'Parque das Hort√™ncias': (-21.8053, -48.1284),
    
    'Parque Iga√ßaba': (-21.7880, -48.1580),
    
    'Parque Residencial Damha': (-21.7538, -48.1971),
    
    'Parque Residencial Jatob√°': (-21.7828, -48.1320),
    
    'Parque Residencial Vale do Sol': (-21.7700, -48.2035),
    
    'Quinta do Salto Residence': (-21.8067, -48.2055),
    
    'Residencial Acapulco': (-21.7900, -48.1990),
    
    'Residencial Alamedas': (-21.7780, -48.1208),
    'Residencial Alamedas II': (-21.7780, -48.1208),
    
    'Residencial Cambuy': (-21.7950, -48.2080),
    
    'Residencial Monte Carlo': (-21.7400, -48.1342),
    
    'Residencial Quinta das Oliveiras': (-21.7870, -48.2040),
    
    'Residencial Ravena': (-21.7930, -48.2000),
    
    'Residencial Santa Luzia': (-21.7495, -48.1449),
    
    'S√£o Geraldo': (-21.7834, -48.1780),
    
    'Solidariedade': (-21.7787, -48.1922),
    
    'Vila Aracoara': (-21.8302, -48.1473),
    
    'Vila Bela Vista': (-21.7920, -48.1850),
    
    'Vila DER': (-21.7880, -48.1720),
    
    'Vila Furlan': (-21.7950, -48.1900),
    
    'Vila Independ√™ncia': (-21.7850, -48.1620),
    
    'Vila Jos√© Bonif√°cio': (-21.7895, -48.1789),
    
    'Vila Melhado': (-21.8059, -48.1673),
    
    'Vila Nossa Senhora Aparecida': (-21.8017, -48.1261),
    
    'Vila Nossa Senhora do Carmo': (-21.7980, -48.1820),
    
    'Vila Progresso': (-21.8000, -48.1650),
    
    'Vila Santana': (-21.7888, -48.1905),
    
    'Vila Sedenho': (-21.7920, -48.1760),
    
    'Vila Yamada': (-21.7850, -48.1690),
    
    'Villa Dei Fiori Residencial': (-21.7920, -48.2060),
    
    # ============================================================
    # BAIRROS FORA DO PER√çMETRO URBANO (coordenadas muito diferentes)
    # ============================================================
    'Jardim Portugal': (-21.7800, -48.1900),  # Corrigido (estava em Mogi das Cruzes)
}


# Fun√ß√£o para obter coordenadas (mesma do arquivo anterior)
def obter_coordenadas_bairro(bairro: str) -> tuple:
    """
    Retorna coordenadas (lat, lon) de um bairro
    
    Args:
        bairro: Nome do bairro
    
    Returns:
        (latitude, longitude) ou None se n√£o encontrado
    """
    if not bairro:
        return None
    
    # Limpar nome do bairro
    bairro_limpo = bairro.strip()
    
    # Busca exata
    if bairro_limpo in BAIRROS_ARARAQUARA:
        return BAIRROS_ARARAQUARA[bairro_limpo]
    
    # Busca case-insensitive
    bairro_lower = bairro_limpo.lower()
    for nome, coords in BAIRROS_ARARAQUARA.items():
        if nome.lower() == bairro_lower:
            return coords
    
    # Busca parcial (cont√©m)
    for nome, coords in BAIRROS_ARARAQUARA.items():
        if bairro_lower in nome.lower() or nome.lower() in bairro_lower:
            return coords
    
    return None


def adicionar_coordenadas_ao_dataframe(df, coluna_bairro='bairro'):
    """
    Adiciona colunas de latitude e longitude ao DataFrame
    """
    import pandas as pd
    
    def get_coords(bairro):
        coords = obter_coordenadas_bairro(bairro)
        if coords:
            return pd.Series({'latitude': coords[0], 'longitude': coords[1]})
        else:
            return pd.Series({'latitude': None, 'longitude': None})
    
    # Aplicar fun√ß√£o
    df[['latitude', 'longitude']] = df[coluna_bairro].apply(get_coords)
    
    # Estat√≠sticas
    total = len(df)
    com_coords = df['latitude'].notna().sum()
    sem_coords = total - com_coords
    
    print(f"\nüìä Geocodifica√ß√£o:")
    print(f"  Total de registros: {total}")
    print(f"  Com coordenadas: {com_coords} ({com_coords/total*100:.1f}%)")
    print(f"  Sem coordenadas: {sem_coords} ({sem_coords/total*100:.1f}%)")
    
    # Mostrar bairros n√£o encontrados
    if sem_coords > 0:
        bairros_nao_encontrados = df[df['latitude'].isna()][coluna_bairro].unique()
        print(f"\n‚ö†Ô∏è Bairros n√£o encontrados ({len(bairros_nao_encontrados)}):")
        for bairro in sorted(bairros_nao_encontrados)[:15]:
            print(f"  - {bairro}")
        
        if len(bairros_nao_encontrados) > 15:
            print(f"  ... e mais {len(bairros_nao_encontrados) - 15}")
    
    return df


def listar_bairros_disponiveis():
    """Lista todos os bairros cadastrados"""
    print(f"Total de bairros cadastrados: {len(BAIRROS_ARARAQUARA)}\n")
    
    # Agrupar por zona
    zonas = {
        'Zona Central': [],
        'Zona Norte (Vila Xavier)': [],
        'Zona Sul': [],
        'Zona Leste': [],
        'Zona Oeste': [],
        'Zona Industrial': [],
        'Outros': []
    }
    
    for bairro in sorted(BAIRROS_ARARAQUARA.keys()):
        if 'Centro' in bairro:
            zonas['Zona Central'].append(bairro)
        elif 'Vila Xavier' in bairro or bairro in ['Vila Biagioni (Vila Xavier)', 'Jardim Brasil (Vila Xavier)']:
            zonas['Zona Norte (Vila Xavier)'].append(bairro)
        elif bairro in ['Vila Harmonia', 'Jardim Higien√≥polis', 'Jardim Bandeirantes', 'Jardim das Flores', 'Jardim dos Manac√°s', 'Jardim Tropical', 'Jardim Nova Araraquara']:
            zonas['Zona Sul'].append(bairro)
        elif bairro in ['Parque Laranjeiras', 'Jardim Imperador', 'Jardim S√£o Francisco', 'Parque Residencial Iguatemi']:
            zonas['Zona Leste'].append(bairro)
        elif bairro in ['Cidade Jardim', 'Vila Ferrovi√°ria', 'Vila Suconasa', 'Campus ville', 'Parque S√£o Paulo', 'Selmi Dei']:
            zonas['Zona Oeste'].append(bairro)
        elif 'Industrial' in bairro or 'Industri√°rios' in bairro:
            zonas['Zona Industrial'].append(bairro)
        else:
            zonas['Outros'].append(bairro)
    
    for zona, bairros in zonas.items():
        if bairros:
            print(f"{zona} ({len(bairros)} bairros):")
            for bairro in bairros:
                coords = BAIRROS_ARARAQUARA[bairro]
                print(f"  {bairro}: {coords}")
            print()


if __name__ == "__main__":
    print("="*60)
    print("BANCO DE COORDENADAS - ARARAQUARA/SP")
    print("="*60)
    listar_bairros_disponiveis()
    
    # Teste de busca
    print("\n" + "="*60)
    print("TESTE DE BUSCA")
    print("="*60)
    
    testes = [
        'Campus ville',
        'Vila Cidade Industrial (Vila Xavier)',
        'Jardim Higien√≥polis',
        'Jardim Residencial Elvio Lupo',
        'Vila Suconasa',
        'Jardim √Årtico'
    ]
    
    for bairro in testes:
        coords = obter_coordenadas_bairro(bairro)
        if coords:
            print(f"‚úì {bairro}: {coords}")
        else:
            print(f"‚úó {bairro}: N√ÉO ENCONTRADO")
            
```

```python
import folium
import math

# Mapear coordenadas para cada bairro
coords = dados['localizacao'].map(BAIRROS_ARARAQUARA)

# Criar lista de im√≥veis apenas para os que t√™m coordenadas
imoveis = []
for idx, row in dados.iterrows():
    if pd.notna(coords.iloc[idx]):
        lat, lon = coords.iloc[idx]
        imoveis.append({
            'nome': row['localizacao'],
            'preco': float(row['preco']),
            'lat': lat,
            'lon': lon,
        })

print(f"Total de im√≥veis com coordenadas: {len(imoveis)}")

# Criar mapa centrado em Araraquara
mapa = folium.Map(location=[-21.7844, -48.1756], zoom_start=13)

# Espalhar im√≥veis que compartilham a mesma coordenada (jitter)
offset_base = 0.0002
contadores = {}
for imovel in imoveis:
    chave = (imovel['lat'], imovel['lon'])
    idx = contadores.get(chave, 0)
    contadores[chave] = idx + 1
    if idx > 0:
        angulo = idx * 0.8
        raio = offset_base * (1 + (idx // 8))
        imovel['lat'] += math.cos(angulo) * raio
        imovel['lon'] += math.sin(angulo) * raio

# Adicionar marcador para cada im√≥vel
for imovel in imoveis:
    if imovel['preco'] < 300000:
        cor = 'green'
    elif imovel['preco'] < 600000:
        cor = 'blue'
    else:
        cor = 'red'
    
    folium.Marker(
        location=[imovel['lat'], imovel['lon']],
        popup=f"<b>{imovel['nome']}</b><br>R$ {imovel['preco']:,.2f}",
        icon=folium.Icon(color=cor, icon='home')
    ).add_to(mapa)

mapa.save('imoveis_por_preco.html')
print("Mapa salvo em 'imoveis_por_preco.html'")
mapa
```
