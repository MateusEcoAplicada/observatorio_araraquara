```python
import pandas as pd
```

```python
df = pd.read_csv('arquivo_viva_real54pgs.csv')
```

```python
df.head()
```

```python
df.drop(columns=['next_page', 'next_page_url'], inplace=True)
```

```python
df["area_m2"] = df["titulo_info"].str.extract(r"Tamanho do im[oó]vel\s+([\d\-]+)\s*m²")
df["quartos"] = df["titulo_info"].str.extract(r"Quantidade de quartos\s+(\d+)")
df["banheiros"] = df["titulo_info"].str.extract(r"Quantidade de banheiros\s+(\d+)")
df["vagas"] = df["titulo_info"].str.extract(r"Quantidade de vagas de garagem\s+(\d+)")

# opcional: converter para numero quando fizer sentido
df["quartos"] = pd.to_numeric(df["quartos"], errors="coerce")
df["banheiros"] = pd.to_numeric(df["banheiros"], errors="coerce")
df["vagas"] = pd.to_numeric(df["vagas"], errors="coerce")

df.head()
```

```python
# Extrair localização após "em"
df["localizacao"] = df["titulo_name"].str.extract(r"em\s+(.*?)$")

df[["titulo_name", "localizacao"]].head(10)
```

```python
df.head()
```

```python
df["imovel"] = df["titulo_name"].str.extract(r"(\w+)\s+para")

df[["titulo_name", "imovel"]].head(10)
```

```python
df.head()
```

```python
df.drop(columns=['titulo_info', 'titulo_info_url', 'titulo_preco_url'], inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=['titulo_name'], inplace=True)
```

```python
df.rename(columns={'titulo_preco': 'preco'}, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df.head()
```

```python
# Remover "R$" e converter espaços para ponto
df['preco'] = df['preco'].str.replace('R$', '', regex=False).str.strip()
df['preco'] = df['preco'].str.replace('.', '', regex=False, n=-1)  # Remove pontos de milhares
df['preco'] = df['preco'].str.replace(',', '.', regex=False)  # Substitui vírgula por ponto
df['preco'] = pd.to_numeric(df['preco'], errors='coerce')
df['area_m2'] = pd.to_numeric(df['area_m2'], errors='coerce')

# Agora remove NaN após conversão (apenas linhas com preço inválido)
df.dropna(subset=['preco'], inplace=True)

df.head()
df.info()
```
